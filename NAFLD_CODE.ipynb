```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score

# Load dataset
df = pd.read_csv(r'C:\Users\adity\OneDrive\Documents\project\NAFLD\resampled_ILPD.csv')

# Check for the 'gender' column
if 'gender' in df.columns:
    # Encode 'Gender' column if it exists
    label_encoder = LabelEncoder()
    df['gender'] = label_encoder.fit_transform(df['gender'])

# Ensure the target variable has only 0 and 1
df['is_patient'] = df['is_patient'].map({1: 0, 2: 1})

# Handle missing values
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# Assign weights based on the new analysis
weights = {
    'direct_bilirubin': -1.0,  # High negative weight
    'tot_bilirubin': -1.0,     # High negative weight
    'tot_proteins': 1.0,       # High positive weight
    'albumin': -1.0,           # High negative weight
    'ag_ratio': -0.5,          # Moderate negative weight
    'sgpt': 1.0,               # High positive weight
    'sgot': 1.0,               # High positive weight
    'alkphos': 1.0             # High positive weight
}

# Apply weights to the features
for feature, weight in weights.items():
    df_imputed[feature] *= weight

# Remove 'tot_bilirubin' from the feature set for the XGBoost model
df_imputed_xgb = df_imputed.drop(columns=['tot_bilirubin'])

# Preprocessing
scaler = MinMaxScaler()

# Preprocessing for models that include 'tot_bilirubin'
X = scaler.fit_transform(df_imputed.iloc[:, :-1].values)
y = df_imputed.iloc[:, -1].values

# Preprocessing for XGBoost without 'tot_bilirubin'
X_xgb = scaler.fit_transform(df_imputed_xgb.iloc[:, :-1].values)

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)
x_train_xgb, x_test_xgb, y_train_xgb, y_test_xgb = train_test_split(X_xgb, y, test_size=0.1, stratify=y, random_state=42)

# Define models
models = {
    'XGB': XGBClassifier(),
    'Support Vector Machine': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'kNN': KNeighborsClassifier(),
    'SGD': SGDClassifier(),
}

# Define hyperparameter grids
param_grids = {
    'XGB': {
        'n_estimators': [100, 200, 500],
        'max_depth': [3, 6, 10],
        'learning_rate': [0.01, 0.1, 0.2]
    },
    'Support Vector Machine': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf', 'sigmoid'],
        'gamma': ['scale', 'auto']
    },
    'Decision Tree': {
        'max_depth': [10, 20, 50, 100],
        'min_samples_split': [2, 5, 10]
    },
    'Random Forest': {
        'n_estimators': [100, 200, 500],
        'max_depth': [10, 20, 50, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'kNN': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
    },
    'SGD': {
        'loss': ['hinge', 'log', 'modified_huber'],
        'penalty': ['l2', 'l1', 'elasticnet'],
        'alpha': [0.0001, 0.001, 0.01]
    }
}

# Function for tuning models with Grid Search
def tune_model_grid_search(model, param_grid, x_train, y_train):
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)
    grid_search.fit(x_train, y_train)
    best_model = grid_search.best_estimator_
    print("Best parameters for {}: {}".format(type(model).__name__, grid_search.best_params_))
    return best_model

# Function for tuning models with Random Search
def tune_model_random_search(model, param_grid, x_train, y_train):
    random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=100, cv=5, scoring='accuracy', verbose=2, n_jobs=-1, random_state=42)
    random_search.fit(x_train, y_train)
    best_model = random_search.best_estimator_
    print("Best parameters for {}: {}".format(type(model).__name__, random_search.best_params_))
    return best_model

# Fine-tune models using Grid Search and Random Search
best_models = {}
for model_name in models:
    print(f"Tuning {model_name} with Grid Search...")
    if model_name == 'XGB':
        best_models[f"{model_name}_grid_search"] = tune_model_grid_search(models[model_name], param_grids[model_name], x_train_xgb, y_train_xgb)
    else:
        best_models[f"{model_name}_grid_search"] = tune_model_grid_search(models[model_name], param_grids[model_name], x_train, y_train)
    
    print(f"Tuning {model_name} with Random Search...")
    if model_name == 'XGB':
        best_models[f"{model_name}_random_search"] = tune_model_random_search(models[model_name], param_grids[model_name], x_train_xgb, y_train_xgb)
    else:
        best_models[f"{model_name}_random_search"] = tune_model_random_search(models[model_name], param_grids[model_name], x_train, y_train)

# Evaluate models
print("Test Set Prediction:\n")
for model_name, model in best_models.items():
    print('-' * 20 + model_name + '-' * 20)
    if "XGB" in model_name:
        y_pred = model.predict(x_test_xgb)
    else:
        y_pred = model.predict(x_test)
    
    print(classification_report(y_test, y_pred, digits=5))
    cf_matrix = confusion_matrix(y_test, y_pred)
    ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    ax.xaxis.set_ticklabels(['False', 'True'])
    ax.yaxis.set_ticklabels(['False', 'True'])
    plt.show()
```
